```python
# ---- 2.2 Outlier Capping (IQR method) ----
def cap_outliers(df, columns):
    df_capped = df.copy()
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_capped[col] = df[col].clip(lower, upper)
    return df_capped

X_capped = cap_outliers(X_imputed, X_imputed.columns)
```
